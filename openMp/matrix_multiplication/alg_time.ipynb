```python
import matplotlib.pyplot as plt
```

```python
#размер матрицы
x = [2, 3, 10, 50, 100, 500, 750, 1000, 1500, 2000]
#время работы обычного алгоритма
t1 = [0.000306, 0.000439, 0.000361, 0.000793, 0.003820, 0.281830, 0.993644, 2.472554, 8.079433, 12.910463]
#время работы оптимизированного алгоритма
t2 = [0.000005, 0.000006, 0.000006, 0.000249, 0.001651, 0.187502, 0.605317, 1.555491, 2.386374, 5.690984]
```

```python
plt.plot(x, t1, label='Обычный')
plt.plot(x, t2, label='Оптимизированный')

plt.xlabel('размер матрицы')
plt.ylabel('время работы алгоритма')
plt.title('Сравнение времени работы алгоритмов')
plt.legend()

plt.show()
```
